```python
import hashlib

class Usuario:
    def __init__(self, nome, sobrenome, email, senha):
        self.__nome = nome
        self.__sobrenome = sobrenome
        self.__email = email
        self.__senha = hashlib.sha256(senha.encode()).hexdigest()

    def nome_completo(self):
        print(self.__nome + " " + self.__sobrenome)

    def checa_senha(self, senha):
        return hashlib.sha256(senha.encode()).hexdigest() == self.__senha
```

```python
u = Usuario("Ana", "Silva", "ana@email.com", "1234")

print(u.nome_completo())
print(u.checa_senha("1234"))
print(u.checa_senha("errada"))
```
