```python
def batata_quente(nomes, MAX):
    # Cria fila para simular roda
    fila = Queue()

    # Coloca os N nomes na fila
    for nome in nomes:
        fila.enqueue(nome)
        
    print("🎮 Início do jogo:")
    fila.print_queue()
    
    # Inicia a lógica do jogo
    while fila.size() > 1:
        
        print(f"\n🔥 Rodada com {MAX} passagens...")
        
        # Simula MAX passagens da batata
        for i in range(MAX):
            fila.enqueue(fila.dequeue())
        
        # Quem estiver na frente após as passagens é eliminado
        eliminado = fila.dequeue()
        print(f"💥 Eliminado: {eliminado}")
        
        fila.print_queue()
    
    # Último restante é o vencedor
    vencedor = fila.dequeue()
    print(f"\n🏆 Vencedor: {vencedor}")
    
    return vencedor
```

```python
nomes = ["Agnaldo", "Bruno", "Patricia", "Humberto", "Renan"]

vencedor = batata_quente(nomes, MAX=3)
print("\nResultado final:", vencedor)
```
